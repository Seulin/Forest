```python
def getKeyword_Tokens(keyword, tokens):
    #첫번째로 등장하는 키워드의 토큰묶음
    nested = [] #next keyword appearnace
    index = [0, 0] #index of tokens
    keyword_pos = 0
    checkOn = False
    cur_pos = 0 #indent position
    for i, tok in enumerate(tokens):
        if tokenize.tok_name[tok.type] == 'NAME' and tok.string == keyword:
            if not checkOn:
                checkOn = True
                keyword_pos = cur_pos
                index[0] = i
            else:
                nested = tokens[i:]
        elif checkOn and tokenize.tok_name[tok.type] == "INDENT":
            #pre_pos = cur_pos
            cur_pos += 1
        elif checkOn and tokenize.tok_name[tok.type] == "DEDENT":
            #pre_pos = cur_pos
            cur_pos -= 1
            if cur_pos == keyword_pos: #comeback
                checkOn = False
                index[1] = i
                return tokens[index[0]:index[1]]
#         elif checkOn and tokenize.tok_name[tok.type] in ["NEWLINE", 'NL']
#             checkOn += 1

            
import suji

def keyword_Length(keyword, tokens):
    
    while tokens
        target    = getKeyword_Tokens(keyword, tokens)
    
        suji.Total_Line(target)

def def_Length(tokens):
    return keyword_Length('def', tokens)

def for_Length(tokens):
    return keyword_Length('for', tokens)

def if_Length(tokens):
    return keyword_Length('if', tokens)
                    
```

```python
from anytree import Node, RenderTree, PreOrderIter


def check_structure(tokens):
        root = Node('root', index = [0, len(tokens) - 1], pos = -1)
        keyword = ['if', 'elif', 'else', 'while', 'for', 'def', 'class'] #keywords demanding indent
        stack = [root]
        cur_pos = 0
        
        def parenting(stack, index):
            node = stack.pop()
            node.index[1] = index
            node.parent = stack[-1]
        
        for (i, tok) in enumerate(tokens):
            if tokenize.tok_name[tok.type] == "NAME" and tok.string in keyword:
                node = Node(tok.string, index = [i, -1], pos = cur_pos)
                stack.append(node)
            elif tokenize.tok_name[tok.type] == "INDENT":
                cur_pos += 1
            elif tokenize.tok_name[tok.type] == 'DEDENT':
                
                cur_pos -= 1
                if stack[-1].pos == cur_pos:
                    parenting(stack, i)
                else:
                    pass
#                     print(stack[-1].pos, cur_pos)
#                     print('else case, not printed')

                # ex) if True: print(30)
            elif tokenize.tok_name[tok.exact_type] == 'COLON' and \
                tokens[i+1].type != 4: #means NEWLINE
                if stack[-1].pos == cur_pos:
                    parenting(stack, i)
                
                # [i for i in range(3)] @@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@this case must be added
                
        return root
                
```
